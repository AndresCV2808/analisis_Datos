```python
from folium.plugins import MiniMap 
from branca.element import Figure

popuptext='<b>Heredia</b>'

my_map = folium.Map(location = [9.974512616110562,-84.1289673943069], zoom_start=16, tiles='stamenterrain')

folium.Marker(location = [9.974512616110562,-84.1289673943069], popup=popuptext, icon=folium. Icon (color="orange ", icon="ok-sign")).add_to (my_map)

folium.Circle(location = [9.974512616110562,-84.1289673943069],color="purple", weight=6, fill_color="red", fill_opacity=0.5, tooltip="Heredia").add_to (my_map)
minimap = MiniMap(tile_layer='stamenterrain', position='bottomright')

my_map.add_child (minimap)

my_map.save('C: folium.html')

my_map

fig = Figure (width=1000, height=500) 
fig.add_child (my_map)
fig
```
